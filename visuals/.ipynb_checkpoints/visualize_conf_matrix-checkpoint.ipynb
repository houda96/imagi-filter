```python
from load_data import load_image, get_data
import torch
import torch.nn as nn
from PIL import Image
import glob, os
from torchvision import transforms
```

```python
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

MODELS = {"lenet": 1, "vgg": 2, "resnet": 3}
TRANSFORMS = {1: transforms.Compose([transforms.ToTensor(), normalize]),
              2: transforms.Compose([transforms.RandomResizedCrop(224), transforms.RandomHorizontalFlip(),
                transforms.ToTensor(), normalize]),
              3: transforms.Compose([transforms.RandomResizedCrop(224), transforms.RandomHorizontalFlip(),
                transforms.ToTensor(), normalize])}
```

```python
validation = "ImageFilteringData/train_test_divisions/val_load_3000.json"
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
```

```python
import json
import numpy as np

options = []
numbering = 0
nummies = [100, 2700, 3500, 700, 4000, 1900, 2600, 4600, 4800, 2600, 2900, 2000, 2000, 3900, 3200, 4400, 4700, 4300]

for m in ['lenet', 'resnet', 'vgg']:
    for c in ['sketches', 'maps', 'icons', 'graphs', 'flags', 'black']:
        options.append((m, c, 'trained_models_finegrained/' + c + '/' + m + '_test_' + 
                        str(nummies[numbering]) + '.pt'))
        numbering += 1
        
        
```

```python
options
```

```python
# Load in chosen model
from sklearn.metrics import confusion_matrix
for option in options:
    model = option[0]
    category = option[1]
    linkie = option[2]
    with open('splits/' + category + '_val_ones.json', 'r') as f:
        val_ones = json.load(f)

    with open('splits/' + category + '_val_zeros.json', 'r') as f:
        val_zeros = json.load(f)

    half_batch = int(40/2)
    val_load = val_ones + val_zeros
    np.random.shuffle(val_load)
    val_load = [val_load[i:i + 40] for i in range(0, len(val_load), 40)]
    
    if model == "lenet":
        net = LeNet(300).to(device)
        net.load_state_dict(torch.load(linkie))
        net.eval()
        resize = True
        transform = TRANSFORMS[1]
    elif model == "resnet":
        net = ResNet152().to(device)
        net.load_state_dict(torch.load(linkie))
        net.eval()
        resize = False
        transform = TRANSFORMS[3]
    elif model == "vgg":
        net = VGG19_BN().to(device)
        net.load_state_dict(torch.load(linkie))
        net.eval()
        resize = False
        transform = TRANSFORMS[2]
        
    targetsA = []
    predictionsA = []

    for batch_id, batch in enumerate(val_load):
        inputs = torch.stack(tuple([load_image(i[0], 300, 300, resize, transform) for i in batch]), 0).to(device)
        targets = torch.LongTensor([i[1] for i in batch]).to(device)
        outputs = net(inputs)
        targetsA.extend(targets.tolist())
        predictionsA.extend(torch.max(outputs, dim=1).indices.tolist())
        
    print(category, model)
    ccc = confusion_matrix(targetsA, predictionsA)
    ccc = (ccc.T / ccc.astype(np.float).sum()).T
    print(ccc)
```
